## [2] Q로 시작하는 부분 처리하기

```python
Contents = [[idx, re.sub("\n", "@@@", con)] for idx, con in enumerate(df.Content.tolist())]
```

```python
df['UpdateCompleteCheck'] = [False]*(len(df))
df['NewContent'] = [[idx, re.sub("\n", "@@@", con)] for idx, con in enumerate(df.Content.tolist())]
```

```python
NewContent = df.NewContent.tolist()
UpdateCompleteCheck = df.UpdateCompleteCheck.tolist()
```

```python
Q = []

for con in Contents:
    if con[1][0].strip() == 'Q':
        Q.append([con[0], con[1].replace('①', '').replace('②', '').replace('③', '').replace('④', '').replace('⑤', '').replace('⑥', '')])
```

```python
print(len(Q))
```

```python
for q in Q:
    print(q[1][:80])
```

```python
def Q_process():
    for i in range(len(Q)):
        idx, con = Q[i]
        if type(con) == dict: continue
        con = "@@@"+con
        qa_dict = dict()
        questions = [q for q in re.findall("@@@[ ]?Q[.:?, '\d)-]+?@*(.*?)\@@@", con)]
        for q in questions:
            if q == "":
                print(idx, con)
        answers = re.sub("@@@[ ]?Q[.:?, '\d)-]+?@*(.*?)\@@@", '###############', con).split("###############")[1:]

        for q, a in zip(questions, answers):
            qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()

        print(questions)

        if input()=='n' or len(questions) == 0 or sum([q for q in questions if type(q)==list]) >= 1 or (len(questions) > len(answers)) or (len(questions) < len(answers)): 
            print(con)
            check = input("Continue?")
            if check == "y" : continue
            elif check == "stop": return i
            else:
                while True:
                    q = input("ENTER Question : ")
                    if q == "n": break
                    elif q == "stop" : return Q, i
                    a = input("ENTER Answer : ")
                    if a == "n": break
                    elif a == "stop" : return Q, i
                    qa_dict[q.replace('@@@',' ')] = a.replace('@@@',' ')
        else:
            for q, a in zip(questions, answers):
                qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
        Q[i] = [idx, qa_dict]
    return
```

```python
result = Q_process()
```

```python
for idx, q in Q:
    NewContent[idx] = q
    UpdateCompleteCheck[idx] = True
```

```python
df['UpdateCompleteCheck'] = UpdateCompleteCheck
df['NewContent'] = NewContent
```

```python
def Num_process():
    for i, start in enumerate(Start_num):
        idx, con = start

        if type(con) == dict: continue
        con = "@@@"+con
        con = con.replace('①', '1 .').replace('②', '2 .').replace('③', '3 .').replace('④', '4 .').replace('⑤', '5 .').replace('⑥', '6 .')
        con = con.replace('⓵', '1 .').replace('⓶', '2 .').replace('⓷', '3 .')
        qa_dict = dict()

        questions = re.findall("@@@\d[가-힣+. ),-]?[. ),\d]+(.*?)\@@@", con)
        answers = re.sub("@@@\d[가-힣+. ),-]?[. ),\d]+(.*?)\@@@", '###############', con).split("###############")[1:]
        
        if len(questions) == 0 or sum([q for q in questions if type(q)==list]) >= 1 or (len(questions) > len(answers)) or (len(questions) < len(answers)):
            questions = re.findall("@@@\d([가-힣+. )*-]+(.*?))\@@@", con)
            answers = re.sub("@@@\d([가-힣+. )*-]+(.*?))\@@@", '###############', con).split("###############")[1:]

        #print(questions)

        if len(questions) == 0 or sum([q for q in questions if type(q)==list]) >= 1 or (len(questions) > len(answers)) or (len(questions) < len(answers)): # check 용도 : input()=='n' or 
            print(con)
            print(questions)
            check = input("Change Content?")
            if check == "stop" : return i
            else:
                while True:
                    q = input("ENTER Question : ")
                    if q == "n": break
                    elif q == "stop" : return
                    a = input("ENTER Answer : ")
                    if a == "n": break
                    elif a == "stop" : return
                    qa_dict[q.replace('@@@',' ')] = a.replace('@@@',' ')
        else:
            for q, a in zip(questions, answers):
                qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
                
        Start_num[i] = [idx, qa_dict]
    return
```

```python
# 전처리 먼저
for idx, con in enumerate(NotUpdateContent):
    temp = con[1].replace('①', '').replace('②', '').replace('③', '').replace('④', '').replace('⑤',  '').replace('⑥', '')
    temp = temp.replace('\u3000', ' ')
    temp = re.sub("@@@Essay+\d", " ", temp)
    temp = re.sub("(@@@취미.*?\@@@.*?)\@@@", "", temp)
    temp = re.sub("(@@@특기.*?)\@@@", "", temp)
    temp = re.sub("(@@@존경인물/이유.*?)\@@@.*?\@@@", "", temp)
    temp = temp.replace('[필수]', ' ')
    temp = temp.replace('@@@직무수행계획', '@@@- 직무수행계획 -')
    temp = re.sub(' +', ' ', temp)
    NotUpdateContent[idx] = [con[0], temp]
```

```python
# 첫 번째 문장 처리
for i, cons in enumerate(NotUpdateContent):
    idx, con = cons
    con = "@@@"+con
    qa_dict = dict()
    questions = re.findall("@@@([A-Za-z0-9가-힣 ,.?+-=`'\"{}()*■□※ⅠⅡⅢⅣ•\[\]<>]+?(시오|인가요|주세요|바랍니다|있나요|하세요|니까|에\s?대하여))+\.?\??", con)
    if not questions:
        questions = [" ".join(q[:1]) for q in re.findall("\[\d.*?\d?\].*?@@@", con)]
        answers = re.sub('\[\d.*?\d?\].*?@@@', '###############', con).split("###############")[1:]
    else:
        questions = [" ".join(q[:1]) for q in questions]
        answers = re.sub("@@@([A-Za-z0-9가-힣 ,.?+-=`'\"{}()*■□※ⅠⅡⅢⅣ•\[\]<>]+?(시오|인가요|주세요|바랍니다|있나요|하세요|니까))+\.?\??", '###############', con).split("###############")[1:]
    if len(questions) == 0 or sum([q for q in questions if type(q)==list]) >= 1 or (len(questions) > len(answers)) or (len(questions) < len(answers)): NotUpdateContent[i] = [idx, {}]
    else:
        for q, a in zip(questions, answers):
            qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
        NotUpdateContent[i] = [idx, qa_dict]
```

```python
def edit(Content, start):
    for i, content in enumerate(Content):
        idx, con = content
        if type(con) == dict and con: continue
        else: con = re.sub("\n", "@@@", NewContent[idx])
        con = "@@@"+con
        qa_dict = dict()
        
        if "자유 양식" in con or "자유양식" in con:
            con = re.sub("@@@자유\s?양식+.*?\@@@", "", con)
            qa_dict['X'] = con
            Content[i] = [idx, qa_dict]
            continue
            
        questions1 = re.findall(patterns[0], con)
        questions = [q[0] for q in questions1]
        if not questions:
            Content[i] = [idx, dict()]
        else:
            answers = re.sub(patterns[0], "", con).split("###############")
            if len(questions) != len(answers): Content[i] = [idx, dict()]
            else:
                print(f"[idx : {i}]")
                print(con)
                print(questions)
                if input("Edit?") == 'n':
                    for q, a in zip(questions, answers):
                        qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
                    Content[i] = [idx, qa_dict]
                    clear_output(wait=True)
                else:
                    while True:
                        question = input("Enter questions : ")
                        if question == 'pass' : Content[i] = [idx, dict()]
                        elif question == 'stop': 
                            Content[i] = [idx, qa_dict]
                            break
                        answer = input("Enter answers : ")
                        qa_dict[question.replace('@@@',' ')] = re.sub(' +', ' ', answer.replace('@@@',' ')).strip()
                    clear_output(wait=True)
    return i, Content
                
```

```python
def edit2(cmd):
    for i, content in enumerate(NotUpdateContent3):
        idx, con = content
        if type(con) == dict: 
            #clear_output(wait=True)
            continue

        qa_dict = {}
        con = "@@@"+con

        print(f"[idx:{i}]")
        print(con)
        questions = re.findall(pattern2[cmd], con)
        print(questions)
        
        check = input("Edit?")
        if check == 'n': 
            answers = re.sub(pattern2[cmd], "", con).split("###############")
            for q, a in zip(questions, answers):
                qa_dict[q[0].replace('@@@',' ')] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
            NotUpdateContent3[i] = [idx, qa_dict]
            clear_output(wait=True)
        elif check == 'p':
            NotUpdateContent3[i] = [idx, dict()]
            clear_output(wait=True)
            continue
        else:
            while True:
                question = input("ENTER Question : ")
                if question == "n": break
                answer = input("ENTER Answer : ")
                qa_dict[question.replace('@@@',' ')] = re.sub(' +', ' ', answer.replace('@@@',' ')).strip()
            NotUpdateContent3[i] = [idx, qa_dict]
            clear_output(wait=True)
```
