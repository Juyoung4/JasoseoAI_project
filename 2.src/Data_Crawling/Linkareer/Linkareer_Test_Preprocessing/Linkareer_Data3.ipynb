```python
total_content = df.NumUpdate.tolist()
```

```python
total_update = df.NotUpdate.tolist()
```

```python
content = [[idx, total_content[idx]] for idx in df[df['NotUpdate'] == False].NumUpdate.index.tolist()]
```

```python
len(content)
```

### [6] 일반 문장 처리

```python
# 전처리 먼저
for idx, con in enumerate(content):
    temp = re.sub("@@@Essay\s+\d", " ", con[1])
    temp = re.sub("(@@@취미.*?\@@@.*?)\@@@", "", temp)
    temp = re.sub("(@@@특기.*?)\@@@", "", temp)
    temp = re.sub("(@@@존경인물/이유.*?)\@@@.*?\@@@", "", temp)
    temp = temp.replace('[필수]', ' ')
    temp = temp.replace('@@@직무수행계획', '@@@- 직무수행계획 -')
    temp = re.sub(' +', ' ', temp)
    content[idx] = [con[0], temp]
```

```python
for i, cons in enumerate(content):
    idx, con = cons
    con = "@@@"+con
    qa_dict = dict()
    questions = re.findall("@@@([A-Za-z0-9가-힣 ,.?+-=`'\"{}()*]+?(시오|인가요|주세요|바랍니다|있나요|지원동기|생활신조와 가치관|대인관계|직무수행계획|보유능력 및 기술))+\.?\??", con)
    if not questions:
        questions = [" ".join(q[:1]) for q in re.findall("\[\d.*?\d?\].*?@@@", con)]
        answers = re.sub('\[\d.*?\d?\].*?@@@', '###############', con).split("###############")[1:]
    else:
        questions = [" ".join(q[:1]) for q in questions]
        answers = re.sub("@@@([A-Za-z0-9가-힣 ,.?+-=`'\"{}()*]+?(시오|인가요|주세요|바랍니다|있나요|지원동기|생활신조와 가치관|대인관계|직무수행계획|보유능력 및 기술))+\.?\??", '###############', con).split("###############")[1:]
    if len(questions) == 0 or sum([q for q in questions if type(q)==list]) >= 1: content[i] = [idx, qa_dict] 
    elif len(questions) > len(answers) or len(questions) < len(answers):
        for q in questions:
            print(q.replace('@@@',' '))
        for a in answers:
            print(a.replace('@@@',' '))
            
        while True:
            question = input("ENTER Question : ")
            if question == "n": break
            answer = input("ENTER Answer : ")
            if answer == "n": break
            qa_dict[question] = answer
        content[i] = [idx, qa_dict] 
    else:
        for q, a in zip(questions, answers):
            qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
        content[i] = [idx, qa_dict]
```

```python
not_fail = [(idx, con) for idx, con in content if con]
fail = [(idx, con) for idx, con in content if not con]
```

```python
print(len(not_fail))
print(len(fail))
```

```python
for idx, con in not_fail:
    total_content[idx] = con
    total_update[idx] = True
df.NumUpdate = total_content
df.NotUpdate = total_update
```
