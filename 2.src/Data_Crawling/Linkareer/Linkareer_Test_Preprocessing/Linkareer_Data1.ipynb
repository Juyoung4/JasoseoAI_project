### [1-1] Company 정보가 3개 이상

* Company 정보는 원래 [회사]/[직군]/[날짜] 형식이어야 되는데 직군을 길게 적을 경우 ("/") split 시 3개 이상 됨
* 이런 데이터 확인해서 변경하기

```python
Company = df.Company.tolist()
```

```python
Company = [com.split('/') for com in Company]
Company = [[c.strip() for c in com] for com in Company]

three_over = []
over = []
j = 0
for idx, com in enumerate(Company):
    if len(com) > 3: 
        three_over.append([idx, com, j])
        j += 1
        if len(com) > 4:
            over.append([idx, com])
```

```python
len(three_over)
```

```python
len(over)
```

```python
three_over
```

```python
CompanyName = []
CompanyCareer = []
CompanyDate = []

for com in Company:
    if len(com) <= 3:
        CompanyName.append(com[0])
        CompanyCareer.append(com[1])
        CompanyDate.append(com[2])
    else:
        CompanyName.append(com[0])
        CompanyDate.append(com[-1])
        CompanyCareer.append(" ".join(com[1:-1]))
```

```python
df['CompanyName'] = CompanyName
df['CompanyCareer'] = CompanyCareer
df['CompanyDate'] = CompanyDate
```

```python
special = re.compile(r'[^ A-Za-z0-9가-힣+]')
result = set()
for user in users:
    result |= set(special.findall(user))
```

### [3-1] \n 여러개를 1개로 줄이기

```python
import re

content2 = [re.sub("\n", "@@@", c) for c in content]
```

### [3-2] 본문을 (질문 - 내용)으로 나누기

#### [3-2-1] 숫자로 시작하는 질문

```python
start_num = []
```

```python
for idx, con in enumerate(content2):
    if con[0].isdigit(): # 숫자로 시작
        start_num.append([idx, con])
```

```python
len(start_num)
```

```python
print([idx for idx, s in start_num][:6])
```

```python
# 숫자로 시작하는 경우
# ①, ② : 이런 숫자 모두 1,2,3,로 변경하기
new_contents = []
for idx, con in start_num:
    # [1] \d[. ]?[. ]+(.*?)(\n)+ 사용해서 질문 가져오기
    con = "@@@"+con
    con = con.replace('①', '1 .').replace('②', '2 .').replace('③', '3 .').replace('④', '4 .').replace('⑤', '5 .').replace('⑥', '6 .')
    con = con.replace('⓵', '1 .').replace('⓶', '2 .').replace('⓷', '3 .')
    qa_dict = dict()
    questions = [q[0] for q in re.findall('@@@\d[. )-]?[. )\d]+(.*?)(\.|@@@)', con)]
    answers = re.sub('@@@\d[. )-]?[. )\d]+(.*?)(\.|@@@)', '###############', con).split("###############")[1:]

    for q, a in zip(questions, answers):
        qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
    new_contents.append([idx, qa_dict])
```
