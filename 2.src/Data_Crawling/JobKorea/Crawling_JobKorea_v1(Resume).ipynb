```python
import requests
import lxml.html
from bs4 import BeautifulSoup
import time

#https://www.jobkorea.co.kr/starter/passassay/View/237136?Page=1&OrderBy=0&FavorCo_Stat=0&Pass_An_Stat=0
#container > div.stContainer > div.starListsWrap.ctTarget > ul > li:nth-child(1) > div.txBx > p > a

main_url = 'https://www.jobkorea.co.kr'           
base_url = 'https://www.jobkorea.co.kr/starter/PassAssay?FavorCo_Stat=0&Pass_An_Stat=0&OrderBy=0&EduType=0&WorkType=0&isSaved=0&Page='

links_url = []
for i in range(1,340) :
    temp_url = base_url + str(i)
    html_doc = requests.get(temp_url, timeout=15, headers={'User-Agent':'Mozilla/5.0'})
    soup = BeautifulSoup(html_doc.content, 'html.parser')
    
    comps = soup.find('div', id='container').find('ul', {'class': 'selfLists'}).find_all('li')

    time.sleep(0.5)
      
    if comps:
        for comp in comps:
            links_url.append(main_url + comp.find('a')['href'])
         
        
    else:
        print("######## move on to next page  ########")
        break
            
else: print("######## max link (", len(links_url),") ########")           
```

```python
import pickle

with open("data.pickle","wb") as fw:
    pickle.dump(links_url, fw)
```
