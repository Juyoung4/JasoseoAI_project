```python
Start_num = []
for i, content in enumerate(NotUpdateContent):
    idx, con = content
    if con[0].isdigit(): # 숫자로 시작
        Start_num.append([idx, con])
    elif con[0] in ('(', '[', '{', '<', '-') or con[1].isdigit() or con[2].isdigit():
        Start_num.append([idx, con])
```

```python
len(Start_num)
```

```python
[idx for idx, con in Start_num][:20]
```

```python
patterns = [
    "@@@((\(?\[?\<?\{?\-?|)\d([가-힣+. \)*-\>\]]+(.*?)))\@@@",
    "@@@((\(?\[?\<?\{?\-?|)\d[가-힣+. \),-\>\]]?[. ),\d]+(.*?))\@@@",
    "@@@([A-Za-z0-9가-힣 ,.ㆍ中外㈜?+-=`'\"{}()&●*■□※&ⅠⅡⅢⅣ•\[\]<>]+?(시오|인가요|주세요|바랍니다|있나요|하세요|니까|에\s?대하여|써라|작성하라|적합한지))+\.?\??\@@@?"
]
```

```python
def Num_process(num=0):
    for i in range(num, len(Start_num)):
        idx, con = Start_num[i]

        if type(con) == dict: continue
        con = "@@@"+con
        qa_dict = dict()
        
        for cmd in range(3):
            questions = [q[0] for q in re.findall(patterns[cmd], con)]
            answers = re.sub(patterns[cmd], '###############', con).split("###############")[1:]
            
            if len(questions) >= 1 or len(questions) == len(answers): 
                for q, a in zip(questions, answers):
                    qa_dict[q] = re.sub(' +', ' ', a.replace('@@@',' ')).strip()
                Start_num[i] = [idx, qa_dict]
                break
        else:
            print(con)
            if input("Edit") == "n" :
                clear_output(wait=True)
                continue
            else:
                while True:
                    q = input("ENTER Question : ")
                    if q == "n": break
                    elif q == "stop" : return
                    a = input("ENTER Answer : ")
                    if a == "n": break
                    elif a == "stop" : return
                    qa_dict[q.replace('@@@',' ')] = a.replace('@@@',' ')
                Start_num[i] = [idx, qa_dict]
                clear_output(wait=True)
    return
```

```python
def qa_dict_make(cmd, con, idx):
    qa_dict = {}
    count = 1
    key_idx, key, value = 0, "", ""
    for idx in range(len(con)):
        content = con[idx].strip()
        if '[공통]' in content:
            if key_idx != idx:
                qa_dict[key] = value
                key = content
                value = ""
            else:
                key = content
                key_idx = idx
            continue
        if len(content) > 50:
            value += (content+ " ")
            continue
        result = re.findall(patterns[cmd], con[idx])
        if result and result[0][0]:
            if idx !=0 :
                if key == '': key = 'X'
                if key in qa_dict.keys():
                    qa_dict[key+str(count)] = value
                    count += 1
                else:
                    qa_dict[key] = value
            if idx-key_idx == 1: 
                del qa_dict[key]
                key += content
                qa_dict[key] = value
            else: key = content
            key_idx = idx
            value = ""
        else:
            value += (content+ " ")

    if key not in qa_dict.keys():
        qa_dict[key] = value
    if key == '':
        qa_dict = {}
        qa_dict['X'] = value
        
    return qa_dict
```

```python
def edit(Keywords, num=0):
    for i in range(num, len(Keywords)):
        idx, con = Keywords[i]
        if type(con) == dict: continue
        print(f"idx : {i}")
        for cmd in range(2):
            qa_dict = qa_dict_make(cmd, con, idx)
            for key, value in qa_dict.items():
                print(key)
                print(value)
                print()
            check = input("cmd Ok?")
            if check == 'y': 
                Keywords[i] = [idx, qa_dict]
                clear_output(wait=False)
                break
            elif check == 'p':
                clear_output(wait=False)
                break
            elif check == 's': return
            else: clear_output(wait=False)
#         else:
#             if input("Edit?") == 'n':
#                 clear_output(wait=False)
#                 continue
#             else:
#                 print("@@@".join(con))
#                 p_qa_dict = dict()
#                 temp = input("Enter Content : ").split("$$$")
#                 for i in range(0, len(temp), 2):
#                     p_qa_dict[temp[i].replace('@@@', ' ')] = temp[i+1].replace('@@@',' ')
#                 Keywords[i] = [idx, p_qa_dict]
    return Keywords
```
